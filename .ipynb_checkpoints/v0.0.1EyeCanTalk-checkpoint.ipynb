```python
from PIL import Image


eyedetect=cv2.CascadeClassifier('haarcascade_eye.xml')
def stack_mv(mv):
    if len(stack)<4:
        stack.append(mv)
    else:
        stack.clear()
        stack.append(mv)
    return stack
def seq_mv(stack_mv):
    if len(stack_mv)==4:
        if 'right'and 'BLINK' in stack_mv:
            word='Call DOC'
        elif 'BLINK' and 'left' and 'right'in stack_mv:
            word='TOILET YA RABKKK'
        elif 'down' and 'BLINK' in stack_mv :
            word='ATCHOU WATER MCHNI JMEL'
        else:
            word=''
    else:
        return 'NOT ENOUGH EYE MOVEMENT'
    return word

def preprocess_ROI(image):
    dsize = (60, 60)
    aa=img_to_array(image)
    image=array_to_img(aa)
    #resized = cv2.resize(cv2.UMat(image), dsize)
    resized = image.resize((60, 60))
    array=img_to_array(resized)
    expanded_image=np.expand_dims(array, axis=0)
    
    return expanded_image
def extract_ROI(frame):
    gozler=[]
    geteye=eyedetect.detectMultiScale(frame)
    for (x,y,w,h) in geteye:
        gozler.append(frame[y:y+h, x:x+w])
        
        
    return gozler
    
model =tf.keras.models.load_model('model_v6_4_class.h5')
#model.save(r'C:\Users\pc\Desktop\EYE_CAN_TAlK_PROJECT\modelv0_ECT.json')
#model =tf.keras.models.load_model('modelv0_ECT.json')
model.summary()
```

```python
import numpy as np
import cv2
import module as m
import time
def pred_eye(frame):
    CLASS=['DOWN', 'LEFT', 'RIGHT', 'UP', 'OPEN']
    
    gray  = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    height, width = gray.shape
    circleCenter = (int(width/2), 50)
    image, face = m.faceDetector(frame, gray)
    image, PointList = m.faceLandmakDetector(frame, gray, face, False)
    RightEyePoint = PointList[36:42]
    LeftEyePoint = PointList[42:48]
    leftRatio, topMid, bottomMid = m.blinkDetector(LeftEyePoint)
    rightRatio, rTop, rBottom = m.blinkDetector(RightEyePoint)
    blinkRatio = (leftRatio + rightRatio)/2
    eye=extract_ROI(gray)
    if len(eye)!=2:#if the eyes array mafihch zouz 3aynin sema fachla 
        mv=''
    else: #lakan fih zouz 3aynin tebda lkhadma
        if blinkRatio > 4:
            mv='BLINK'
                
        else:
                
            ROI_right=preprocess_ROI(eye[0])
            ROI_left=preprocess_ROI(eye[1])

            m1=CLASS[model.predict_classes(ROI_right)[0]]
            m2=CLASS[model.predict_classes(ROI_left)[0]]

            if m1=='RIGHT':
                mv='right'
            elif m1=='LEFT':
                mv='left'
            elif m1=='UP':
                mv='up'
            elif m1=='DOWN':
                mv='down'
                #elif m1=='OPEN':
                    #mv='NEUTRAL'

            else:
                mv=''
    font = cv2.FONT_HERSHEY_SIMPLEX
    color = (0, 0, 0)
    stroke = 2
    
    return mv
```

```python
import time
import numpy as np
import cv2
import imutils
from imutils.video import VideoStream
from tensorflow import keras

model = keras.models.load_model('modelv0.h5')


def video_mask_detector():
    video = VideoStream(src=0).start()
    time.sleep(1.0)
    while True:
        # Capture frame-by-frame
        frame = video.read()

        mv = pred_eye(frame)
        # Display the resulting frame
        # show the output frame
        font = cv2.FONT_HERSHEY_SIMPLEX
        color = (0, 0, 0)
        stroke = 2
    
        cv2.putText(frame, mv, (50,50), font, 1, color, stroke, cv2.LINE_AA)
        cv2.imshow("Mask detector", frame)

        key = cv2.waitKey(1) & 0xFF
        # if the `q` key was pressed, break from the loop
        if key == ord("q"):
            break
    # cleanup
    cv2.destroyAllWindows()
    video.stop()
```
