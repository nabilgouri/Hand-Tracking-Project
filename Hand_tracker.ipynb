# Imports

```python
import cv2
import numpy as np
import mediapipe as mp
import time
```

# Constants

```python
font = cv2.FONT_HERSHEY_SIMPLEX
org = (50, 50)
fontScale = 1
color = (255, 0, 0)
thickness = 2
```

# Running the webCam

```python
cap=cv2.VideoCapture(0)
mpHands=mp.solutions.hands
hands=mpHands.Hands()
mpDraw=mp.solutions.drawing_utils

previousTime=0



while True:
    ret, img=cap.read()
    imgBGR=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    result=hands.process(imgBGR)
    
    
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks:
        for landmarks in result.multi_hand_landmarks :
            mpDraw.draw_landmarks(img, landmarks, mpHands.HAND_CONNECTIONS)
            
    cv2.imshow('Frame', img)
    currentTime=time.time()
    fps=1-(currentTime-previousTime)
    previousTime=currentTime
    cv2.putText(img, f'FPS: {round(fps*100)}', org, font, fontScale, color, thickness, cv2.LINE_AA)
    
    cv2.imshow('Frame', img)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
